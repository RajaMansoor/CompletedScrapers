```python
for page in range(17808, 89475):
    
    page_url = f"http://htgs.ccgp.gov.cn/GS8/contractpublish/index_{str(page)}"
    print(f"Page Url: {page_url}")
    
    driver.get(page_url)
    time.sleep(0.05)
    
    soup = BeautifulSoup(driver.page_source, "html.parser")
    lis = soup.find("ul", {"class": "ulst"}).find_all("li", {"style":"height:60px"})
    
    for li in lis:
    
        url = li.find("a").get("href").replace(".", "")
        url = f"http://htgs.ccgp.gov.cn/GS8/contractpublish{url}"
        print(url)

        urls.append(url)
        
    print(f"🟤⚫🟠: {len(list(set(urls)))}")
```

```python
contract_numbers = []
contract_names = []
project_numbers = []
project_names = []
purchaser_a_parties = []
addresses = []
contacts = []
supplier_b_parties = []
address_2_numbers = []
contacts_2 = []
main_subject_names = []
number_of_main_targets = []
main_target_unit_prices = []
contract_amounts = []
performance_periods_and_locations = []
purchasing_methods = []
contract_signin_dates = []
contract_announcement_dates = []
```

```python
for index, url in enumerate(urls[28985:30000], start=28985):
    
    print(f"🟧🟨🟧{str(index)}--- {url}")
    driver.get(url)
    
    time.sleep(0.5)
    soup = BeautifulSoup(driver.page_source, "html.parser")
    
    
    try:
        contract_number = soup.find(lambda tag: tag.name == 'strong' and '合同编号' in tag.text).get_text().split("  ")[1].strip()
        print(contract_number)
        contract_numbers.append(contract_number)
    except Exception as e:
        print(f"Got exception in contract number: {e}")
        contract_numbers.append("")
        
    try:
        contract_name = soup.find(lambda tag: tag.name == 'strong' and '合同名称' in tag.text).get_text().split("  ")[1].strip()
        print(contract_name)
        contract_names.append(contract_name)
    except Exception as e:
        print(f"Got exception in contract name: {e}")
        contract_names.append("")
        
    try:
        project_number = soup.find(lambda tag: tag.name == 'strong' and '项目编号' in tag.text).get_text().split("  ")[1].strip()
        print(project_number)
        project_numbers.append(project_number)
    except Exception as e:
        print(f"Got exception in project_number: {e}")
        project_numbers.append("")
        
    try:
        project_name = soup.find(lambda tag: tag.name == 'strong' and '项目名称' in tag.text).get_text().split("  ")[1].strip()
        print(project_name)
        project_names.append(project_name)
    except Exception as e:
        print(f"Got exception in project_name: {e}")
        project_names.append("")
        
    try:
        purchaser_a_party = soup.find(lambda tag: tag.name == 'p' and '采购人（甲方）' in tag.text).get_text().split("  ")[1].strip()
        print(purchaser_a_party)
        purchaser_a_parties.append(purchaser_a_party)
    except Exception as e:
        print(f"Got exception in purchaser_a_party: {e}")
        purchaser_a_parties.append("")
        
    try:
        address = soup.find(lambda tag: tag.name == 'p' and '地  址' in tag.text).get_text().replace("地  址：", "").strip()
        print(address)
        addresses.append(address)
    except Exception as e:
        print(f"Got exception in address: {e}")
        addresses.append("")
        
    try:
        contact = soup.find(lambda tag: tag.name == 'p' and '联系方式' in tag.text).get_text().replace("联系方式：", "").strip()
        print(contact)
        contacts.append(contact)
    except Exception as e:
        print(f"Got exception in contact: {e}")
        contacts.append("")
    
    try:
        supplier_b_party = soup.find(lambda tag: tag.name == 'p' and '供应商（乙方）' in tag.text).get_text().replace("供应商（乙方）：", "").strip()
        print(supplier_b_party)
        supplier_b_parties.append(supplier_b_party)
    except Exception as e:
        print(f"Got exception in supplier_b_party: {e}")
        supplier_b_parties.append("")
    
    try:
        address_2_number = soup.find_all(lambda tag: tag.name == 'p' and '地  址：' in tag.text)[1].get_text().replace("地  址：", "").replace("2号", "").strip()
        print(address_2_number)
        address_2_numbers.append(address_2_number)
    except Exception as e:
        print(f"Got exception in address_2_number: {e}")
        address_2_numbers.append("")
        
        
    try:
        contact = soup.find_all(lambda tag: tag.name == 'p' and '联系方式' in tag.text)[1].get_text().replace("联系方式：", "").strip()
        print(contact)
        contacts_2.append(contact)
    except Exception as e:
        print(f"Got exception in contact: {e}")
        contacts_2.append("")
        
        
    try:
        main_subject_name = soup.find(lambda tag: tag.name == 'p' and '主要标的名称：' in tag.text).get_text().replace("主要标的名称：", "").strip()
        print(main_subject_name)
        main_subject_names.append(main_subject_name)
    except Exception as e:
        print(f"Got exception in main_subject_name: {e}")
        main_subject_names.append("")
    
    try:
        number_of_main_target = soup.find(lambda tag: tag.name == 'p' and '主要标的数量：' in tag.text).get_text().replace("主要标的数量：", "").strip()
        print(number_of_main_target)
        number_of_main_targets.append(number_of_main_target)
    except Exception as e:
        print(f"Got exception in number_of_main_target: {e}")
        number_of_main_targets.append("")
    
    try:
        main_target_unit_price = soup.find(lambda tag: tag.name == 'p' and '主要标的单价：' in tag.text).get_text().replace("主要标的单价：", "").strip()
        print(main_target_unit_price)
        main_target_unit_prices.append(main_target_unit_price)
    except Exception as e:
        print(f"Got exception in main_target_unit_price: {e}")
        main_target_unit_prices.append("")
    
    try:
        contract_amount = soup.find(lambda tag: tag.name == 'p' and '合同金额：' in tag.text).get_text().replace("合同金额：", "").strip()
        print(contract_amount)
        contract_amounts.append(contract_amount)
    except Exception as e:
        print(f"Got exception in contract_amount: {e}")
        contract_amounts.append("")
    
    try:
        performance_periods_and_location = soup.find(lambda tag: tag.name == 'p' and '履约期限、地点等简要信息：' in tag.text).get_text().replace("履约期限、地点等简要信息：", "").strip()
        print(performance_periods_and_location)
        performance_periods_and_locations.append(performance_periods_and_location)
    except Exception as e:
        print(f"Got exception in performance_periods_and_location: {e}")
        performance_periods_and_locations.append("")
        
    try:
        purchasing_method = soup.find(lambda tag: tag.name == 'p' and '采购方式：' in tag.text).get_text().replace("采购方式：", "").strip()
        print(purchasing_method)
        purchasing_methods.append(purchasing_method)
    except Exception as e:
        print(f"Got exception in purchasing_method: {e}")
        purchasing_methods.append("")
    
    try:
        contract_signin_date = soup.find(lambda tag: tag.name == 'strong' and '七、合同签订日期：' in tag.text).get_text().replace("七、合同签订日期：", "").strip()
        print(contract_signin_date)
        contract_signin_dates.append(contract_signin_date)
    except Exception as e:
        print(f"Got exception in contract_signin_date: {e}")
        contract_signin_dates.append("")
    
    try:
        contract_announcement_date = soup.find(lambda tag: tag.name == 'strong' and '八、合同公告日期：' in tag.text).get_text().replace("八、合同公告日期：", "").strip()
        print(contract_announcement_date)
        contract_announcement_dates.append(contract_announcement_date)
    except Exception as e:
        print(f"Got exception in contract_announcement_date: {e}")
        contract_announcement_dates.append("")
    
   
```

```python
output = pd.DataFrame([])

output["Contract Number"] = contract_numbers
output["Contract Name"] = contract_names
output["Project Number"] = project_numbers
output["Project Name"] = project_names
output["Purchaser"] = purchaser_a_parties
output["Address"] = addresses
output["Contact"] = contacts
output["Supplier"] = supplier_b_parties
output["Address 2"] = address_2_numbers
output["Contact 2"] = contacts_2
output["Main Subject Name"] = main_subject_names
output["No of Main Targets"] = number_of_main_targets
output["Main Target Unit Price"] = main_target_unit_prices
output["Contract Amount"] = contract_amounts
output["Performance Period and Location"] = performance_periods_and_locations
output["Purchasing Method"] = purchasing_methods
output["Contract Signin Date"] = contract_signin_dates
output["Contract Announcement Date"] = contract_announcement_dates

output.to_excel("ccgp_3_f.xlsx", index=False)
```
