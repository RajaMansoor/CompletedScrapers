```python
from selenium import webdriver
import pandas as pd
from bs4 import BeautifulSoup
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome(executable_path='./chromedriver')
```

```python
url = 'https://eservices.scottsdaleaz.gov/maps/registered-rentals'
driver.get(url)
```

```python
site_address = []
tax_num = []
owner_name = []
com_name = []
mail_add = []
emails = []
phones = []
mgmts = []
mgmt_company = []
mgmt_contact = []
mgmt_address = []
mgmt_phone = []
mgmt_hrphone = []
mgmt_email = []
dates = []
```

```python
soup = BeautifulSoup(driver.page_source, "html.parser")
td = soup.find("li", {"class": "identify-subitem"}).find_all("td")


try:
    address = td[1].text
    print(address)
    site_address.append(address)
except:
    print(f"Got exception in address: {str(e)}")
    site_address.append(" ")
    
try:
    tax = td[3].text
    print(tax)
    tax_num.append(tax)
except:
    print(f"Got exception in tax: {str(e)}")
    tax_num.append(" ")
        
try:        
    ownername = td[5].text
    print(ownername)
    owner_name.append(ownername)
except:
    print(f"Got exception in ownername: {str(e)}")
    owner_name.append(" ")

try:
    comname = td[7].text
    print(comname)
    com_name.append(comname)
except:
    print(f"Got exception in comname: {str(e)}")
    com_name.append(" ")

try:
    mailadd = td[9].text
    print(mailadd)
    mail_add.append(mailadd)
except:
    print(f"Got exception in mail_add: {str(e)}")
    mail_add.append(" ")
    
try:
    email = td[11].text
    print(email)
    emails.append(email)
except:
    print(f"Got exception in email: {str(e)}")
    emails.append(" ")

try:
    phone = td[13].text
    print(phone)
    phones.append(phone)
except:
    print(f"Got exception in phone: {str(e)}")
    phones.append(" ")
    
try:
    mgmt = td[15].text
    print(mgmt)
    mgmts.append(mgmt)
except:
    print(f"Got exception in mgmt: {str(e)}")
    mgmts.append(" ")
    
try:
    mgmtcom = td[17].text
    print(mgmtcom)
    mgmt_company.append(mgmtcom)
except:
    print(f"Got exception in mgmtcom: {str(e)}")
    mgmt_company.append(" ")
    
try:
    mgmtcont = td[19].text
    print(mgmtcont)
    mgmt_contact.append(mgmtcont)
except:
    print(f"Got exception in mgmtcont: {str(e)}")
    mgmt_contact.append(" ")

try:
    mgmtaddress = td[21].text
    print(mgmtaddress)
    mgmt_address.append(mgmtaddress)
except:
    print(f"Got exception in mgmtaddress: {str(e)}")
    mgmt_address.append(" ")

try:
    mgmtphone = td[23].text
    print(mgmtphone)
    mgmt_phone.append(mgmtphone)
except:
    print(f"Got exception in mgmtphone: {str(e)}")
    mgmt_phone.append(" ")
    
try:
    mgmthrphone = td[25].text
    print(mgmthrphone)
    mgmt_hrphone.append(mgmthrphone)
except:
    print(f"Got exception in mgmthrphone: {str(e)}")
    mgmt_hrphone.append(" ")
    
try:  
    mgmtemail = td[27].text
    print(mgmtemail)
    mgmt_email.append(mgmtemail)
except:
    print(f"Got exception in mgmtemail: {str(e)}")
    mgmt_email.append(" ")

try:    
    date = td[29].text
    print(date)
    dates.append(date)
except:
    print(f"Got exception in date: {str(e)}")
    dates.append(" ")
    

print(len(site_address),len(tax_num),len(owner_name),len(com_name),len(mail_add),len(emails),len(phones),len(mgmts),len(mgmt_company),len(mgmt_contact),len(mgmt_address),len(mgmt_phone),len(mgmt_hrphone),len(mgmt_email),len(dates))
```

```python
data = pd.DataFrame([site_address, tax_num, owner_name, com_name, mail_add, emails, phones, mgmts, mgmt_company, mgmt_contact, mgmt_address, mgmt_phone, mgmt_hrphone, mgmt_email, dates])
data = data.transpose() 
data.columns=["site_address", "tax_num",  "owner_name", "com_name", "mail_add", "emails", "phones", "mgmts", "mgmt_company", "mgmt_contact", "mgmt_address", "mgmt_phone", "mgmt_hrphone", "mgmt_email", "dates"]
data.head(5)
data.to_csv("cmap.csv", index=False)
```
