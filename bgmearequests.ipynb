```python
from bs4 import BeautifulSoup
import pandas as pd
import time
import requests
```

```python
links = []
```

```python
for page in range(169,182):
    page_url = f"https://www.bgmea.com.bd/page/member-list?page={page}"
    print(f"🔴🔴____{page_url}")
    try:
        source = requests.get(page_url)
        time.sleep(0.3)
        soup = BeautifulSoup(source.content, "html.parser")
        urls = soup.find_all("a", {"class": "btn btn-sm btn-info"})
        for url in urls:
            link = url.get("href")
            print(link)
            links.append(link)
    except Exception as error:
        print(f"Got Exception in mobile: {str(error)}")
        break
    

```

```python
output = pd.DataFrame([])
output["links"] = links
output.to_excel("bgmea_links.xlsx", index=False)
```

```python
file = pd.read_excel("bgmea_links.xlsx")
links = list(file["links"])
```

```python
len(links)
```

```python
company_names = []
bgmreg_nos = []
positions = []
names = []
mobiles = []
emails = []
contract_person_1s = []
designation_1s = []
mailing_address_3s = []
factory_address_3s = []
websites = []
dates = []
```

```python
for index,link in enumerate(links[3131:],start = 3131):
    print(f"🚀🚀🚀🚀🚀🚀----------{index}--------{link}")
    source = requests.get(link)
    time.sleep(0.5)
    soup = BeautifulSoup(source.content, "html.parser")
    try:
        company_name = soup.find("thead", {"class": "thead-light"}).find("strong").text.strip()
        print(company_name)
        company_names.append(company_name)
    except Exception as e:
        print(f"Got exception in company_name: {e}")
        company_names.append(" ")

    try:
        company_info = soup.find("div", {"id": "company_info"}).find("tbody")
        bgmreg_no = company_info.find(lambda tag: tag.name == 'th' and "BGMEA Reg. No." in tag.text.strip()).find_next("td").text.strip()
        print(bgmreg_no)
        bgmreg_nos.append(bgmreg_no)
    except Exception as e:
        print(f"Got exception in bgmreg_no: {e}")
        bgmreg_nos.append(" ")
    try:
        tds = company_info.find("tbody").find_all("td")
        try:
            position = tds[0].text.strip()
            print(position)
            positions.append(position)
        except Exception as e:
            print(f"Got exception in position: {e}")
            positions.append(" ")
        try:
            name = tds[1].text
            print(name)
            names.append(name)
        except Exception as e:
            print(f"Got exception in name: {e}")
            names.append(" ")
        try:
            mobile = tds[2].text.strip()
            print(mobile)
            mobiles.append(mobile)
        except Exception as e:
            print(f"Got exception in mobile: {e}")
            mobiles.append(" ")
        try:
            email = tds[3].text.strip()
            print(email)
            emails.append(email)
        except Exception as e:
            print(f"Got exception in email: {e}")
            emails.append(" ")
    except Exception as e:
        print(f"Got exception in ❌❌❌ company_info_tds: {e}")
        positions.append(" ")
        names.append(" ")
        mobiles.append(" ")
        emails.append(" ")
    
    try:
        address_info = soup.find("div", {"id": "address_info"}).find("tbody")
        try:
            contract_person = address_info.find(lambda tag: tag.name == 'th' and "Contact Person" in tag.text.strip()).find_next("td").text
            contract_person_1 = contract_person.strip("Name:").strip( )
            print(contract_person_1)
            contract_person_1s.append(contract_person_1)
        except Exception as e:
            print(f"Got exception in contract_person: {e}")
            contract_person_1s.append(" ")
            
        try:
            designation = address_info.find(lambda tag: tag.name == 'th' and "Contact Person" in tag.text.strip()).find_next("tr").find("td").text
            designation_1 = designation.replace("Designation:", '').strip( )
            print(designation_1)
            designation_1s.append(designation_1)
        except Exception as e:
            print(f"Got exception in designation: {e}")
            designation_1s.append(" ")

        try:
            mailing_address = address_info.find(lambda tag: tag.name == 'th' and "Mailling Address" in tag.text.strip()).find_next("td").contents
            mailing_address_2 =[]
            for i in mailing_address:
                try:
                    mailing_address_1=i.strip( )
                    mailing_address_2.append(mailing_address_1)
                except:
                    pass
            mailing_address_3 =" ".join(mailing_address_2)
            print(mailing_address_3)
            mailing_address_3s.append(mailing_address_3)
        except Exception as e:
            print(f"Got exception in mailing_address: {e}")
            mailing_address_3s.append(" ")

        try:
            factory_address = address_info.find(lambda tag: tag.name == 'th' and "Factory Address" in tag.text.strip()).find_next("td").contents
            factory_address_2 =[]
            for f in factory_address:
                try:
                    factory_address_1=f.strip( )
                    factory_address_2.append(factory_address_1)
                except:
                    pass
            factory_address_3 =" ".join(factory_address_2)
            print(factory_address_3)
            factory_address_3s.append(factory_address_3)
        except Exception as e:
            print(f"Got exception in factory_address: {e}")
            factory_address_3s.append(" ")
    except Exception as e:
        print(f"Got exception in ❌❌❌ address_info: {e}")
        contract_person_1s.append(" ")
        designation_1s.append(" ")
        mailing_address_3s.append(" ")
        factory_address_3s.append(" ")
        
    try:
        final_info = soup.find("div", {"id": "final_info"}).find("tbody")
        try:
            website = final_info.find(lambda tag: tag.name == 'th' and "Website" in tag.text.strip()).find_next("td").find("a").get("href")
            print(website)
            websites.append(website)
        except Exception as e:
            print(f"Got exception in website: {e}")
            websites.append(" ")
        try:
            date = final_info.find(lambda tag: tag.name == 'th' and "Date of Establishment" in tag.text.strip()).find_next("td").text.strip()
            print(date)
            dates.append(date)
        except Exception as e:
            print(f"Got exception in date: {e}")
            dates.append(" ")
    except Exception as e:
        print(f"Got exception in ❌❌❌ final_info: {e}")
        websites.append(" ")
        dates.append(" ")
```

```python

output = pd.DataFrame([])
output["Company Name"] = company_names
output["BGMEA Reg. No."] = bgmreg_nos
output["Position"] = positions
output["Name"] = names
output["Mobile No."] = mobiles
output["Email"] = emails
output["Contact Person Name"] = contract_person_1s
output["Contact Person Designation"] = designation_1s
output["Mailling Address"] = mailing_address_3s
output["Factory Address"] = factory_address_3s
output["Website"] = websites
output["Date of Establishment"] = dates
output.to_excel("bgmea_data_0-3617.xlsx", index=False)
```
