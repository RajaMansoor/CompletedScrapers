```python
#Importing supporting libraries
import random
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.select import Select
from selenium.webdriver.common.by import By
import datetime
import pandas as pd
from bs4 import BeautifulSoup
import math
import datetime
import os
from urllib.parse import urlparse
import requests
```

```python
#Starting Chrome Incognito window to automatically login.
option = webdriver.ChromeOptions()
driver = webdriver.Chrome(executable_path='./chromedriver', options=option)
time.sleep(2)
```

```python
driver.get("https://horoguides.com/hk/brands_list")
```

```python
soup = BeautifulSoup(driver.page_source, "html.parser")
brands_1 = soup.find_all("div", {"class": "unit unitL Ming5 column boxSizing"})
brands_2 = soup.find_all("div", {"class": "unit unitR Ming5 column boxSizing"})
brand_links = []
for brand_1 in brands_1:
    brand_3 = brand_1.find("a").get("href")
    brand_links.append(brand_3)
for brand_2 in brands_2:
    brand_4 = brand_2.find("a").get("href")
    brand_links.append(brand_4)
    
```

```python
product_urls = []
```

```python
import math
for index,brand_link in enumerate(brand_links[75:],start = 75):
    brand_url = f"https://horoguides.com{brand_link}"
    print(f"index:________________{str(index)}________________{brand_url}")
    try:
        driver.get(brand_url)
        time.sleep(10)
        soup_1 = BeautifulSoup(driver.page_source, "html.parser")
        loader = soup_1.find("div", {"class": "showArea"}).find("span").text.strip()
        print(f"loader________:{loader}")
        loader_1 = int(loader)
        loader_count = math.ceil(loader_1/8)
        print(loader_count)
        for i in range(loader_count-1):
            load = driver.find_element(By.ID, "moreBtn_1")
            time.sleep(5)
            load.click()
            time.sleep(7)
        soup_2 = BeautifulSoup(driver.page_source, "html.parser")
        divs = soup_2.find_all("div", {"class": "fWatch"})
        for div in divs:
            a = div.find("a").get("href")
            product_urls.append(a)
            print(a)
        print(len(product_urls))
    except Exception as e:
        print(f"Got exception_________: {e}")
        break
    
        
        
    
    
    
```

```python
output = pd.DataFrame([])
output["product_urls"] = product_urls
output.to_excel("product_url_75.xlsx", index=False)
```
