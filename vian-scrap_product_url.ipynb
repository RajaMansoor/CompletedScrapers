```python
#Importing supporting libraries
import random
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.select import Select
from selenium.webdriver.common.by import By
import datetime
import pandas as pd
from bs4 import BeautifulSoup
import math
import datetime
import os
from urllib.parse import urlparse
import requests
```

```python
#Starting Chrome Incognito window to automatically login.
option = webdriver.ChromeOptions()
driver = webdriver.Chrome(executable_path='./chromedriver', options=option)
time.sleep(2)
```

```python
driver.get("https://www.viantp.com/")
```

```python
file = pd.read_excel("sub_cats_4.xlsx")
sub_sub_cats_2 = list(file["sub_sub_cats_2"])
sub_cats_2 = list(file["sub_cats_2"])
main_cats_2 = list(file["main_cats_2"])
print(len(sub_sub_cats_2))
```

```python
product_links = []
sub_sub_cats_name = []
sub_cats_name = []
main_cats_name = []
```

```python

for sub_sub_link,sub_link,main_link in zip(sub_sub_cats_2,sub_cats_2,main_cats_2):
    url = f"https://www.viantp.com{sub_sub_link}"
    print(len(product_links))
    print(url)
    
    driver.get(url)
    time.sleep(15)
    
    while True:
        soup = BeautifulSoup(driver.page_source, "html.parser")
        divs = soup.find_all("div",{"class": "resultsStatic"})
        for div in divs:
            product_link = div.find("a").get("href")
            print(product_link)
            product_links.append(product_link)

            print(sub_sub_link[21:])
            sub_sub_cats_name.append(sub_sub_link[21:])

            print(sub_link)
            sub_cats_name.append(sub_link)

            print(main_link)
            main_cats_name.append(main_link)
        
        try:
            driver.find_element(By.LINK_TEXT, "Next").click()
            time.sleep(15)
        except:
            break
        
```

```python
output = pd.DataFrame([])

output["product_links"] = product_links
output["sub_sub_cats_name"] = sub_sub_cats_name
output["sub_cats_name"] = sub_cats_name
output["main_cats_name"] = main_cats_name
output.to_excel("product_link_e.xlsx", index=False)
```
